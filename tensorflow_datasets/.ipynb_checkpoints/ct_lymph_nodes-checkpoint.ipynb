```python

import tensorflow_datasets as tfds
import tensorflow as tf

import numpy as np
import os
import io
import pydicom
import nibabel 
#tfds.load('ct_lymph_nodes')
```

```python
ds = tfds.load('ct_lymph_nodes')
```

```python
ds_iter = iter(ds['train'])
ds_iter
```

```python
sample = next(ds_iter)
sample
```

```python
import matplotlib.pyplot as plt
image = sample['image']
mask = sample['mask']
fig, axes = plt.subplots(1,2, figsize=(10,10))
axes[0].imshow(tf.squeeze(image))
axes[0].set_title('Image')
axes[1].imshow(tf.squeeze(mask))
axes[1].set_title('Mask')
plt.suptitle(sample['age'].numpy().decode("utf-8") +' '+ sample['sex'].numpy().decode("utf-8") +' '+sample['body_part'].numpy().decode("utf-8") )
plt.show()
```
